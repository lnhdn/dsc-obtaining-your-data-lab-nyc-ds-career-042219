````python
# Your code here
import sqlite3
import pandas as pd
````

````python
con = sqlite3.connect('lego.db')
c = con.cursor()
c.execute('''SELECT name FROM sqlite_master
             WHERE type='table'
             ORDER BY name''')
c.fetchall()
````

````python
c.execute('''select * from product_details''')
product_details = pd.DataFrame(c.fetchall())
product_details.columns = [i[0] for i in c.description]
product_details.head()
````

````python
c.execute('''select * from product_info''')
product_info = pd.DataFrame(c.fetchall())
product_info.columns = [i[0] for i in c.description]
product_info.head()
````

````python
c.execute('''select * from product_pricing''')
product_pricing = pd.DataFrame(c.fetchall())
product_pricing.columns = [i[0] for i in c.description]
product_pricing.head()
````

````python
c.execute('''select * from product_reviews''')
product_reviews = pd.DataFrame(c.fetchall())
product_reviews.columns = [i[0] for i in c.description]
product_reviews.head()
````
